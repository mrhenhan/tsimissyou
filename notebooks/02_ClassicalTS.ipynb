# Classic TS Analysis
Classic TS analysis methods

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds = pd.Series(np.random.normal(size=500))
```

```python
mpl.rcParams['figure.figsize'] = (15, 10) # use bigger graphs
def func(x):
    return x.iloc[0] - (0.9*x.iloc[1]) + x.iloc[2]

sns.lineplot(data=ds)



#ds.rolling(window=3).mean().plot()
#ds.rolling(window=3, center=True).mean().plot()
ds.rolling(window=3).apply(func).plot()
#ds.expanding(min_periods=3).mean().plot()
#ds.ewm(span=20, adjust=False).mean().plot()
```

```python
ds2 = pd.Series(np.random.normal(size=30)).to_frame()
ds2.head()
```

```python
def func(x):
    return x.iloc[0]-0.9*x.iloc[1] + x.iloc[2]


ds2['rm'] = ds2.rolling(window=3, center=False).apply(func)
ds2[0].plot()
ds2['rm'].plot()
```
